# Projection

This notebook projects detected bounding boxes to the 2D room map by calling `run_projection()`.

```python
from pathlib import Path
from types import SimpleNamespace
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
SRC_DIR = PROJECT_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

import config
from projection import run_projection
```

```python
args = SimpleNamespace(
    input_csv=config.YOLO_DETECTIONS_CSV,
    output_csv=config.PROJECTED_DETECTIONS_CSV,
    human_model_path=config.HUMAN_PROJECTION_MODEL_PATH,
    machine_model_path=config.MACHINE_PROJECTION_MODEL_PATH,
    normalized_factor=config.NORMALIZED_FACTOR,
)
args
```

```python
projected_rows = run_projection(args)
len(projected_rows), projected_rows[:3]
```

```python
summary = {}
for row in projected_rows:
    summary[row['projection_model']] = summary.get(row['projection_model'], 0) + 1
summary
```
