# YOLO Training

This notebook keeps YOLO retraining available. Training is intentionally gated by `RUN_TRAINING` because it is expensive.

```python
from pathlib import Path
from types import SimpleNamespace
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
SRC_DIR = PROJECT_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

import config
from yolo_train import train_yolo
```

```python
RUN_TRAINING = False

args = SimpleNamespace(
    data=config.YOLO_TRAINING_DIR / 'dataset' / 'data.yaml',
    model=config.YOLO_BASE_MODEL_PATH,
    epochs=100,
    imgsz=640,
    batch=16,
    project=config.YOLO_TRAINING_PROJECT_DIR,
    name='cleanroom_yolo',
    device=None,
    workers=8,
    exist_ok=True,
)
args
```

```python
if RUN_TRAINING:
    training_result = train_yolo(args)
else:
    training_result = 'Training not run. Set RUN_TRAINING = True after preparing data.yaml.'
training_result
```
